```python
import matplotlib.pyplot as plt
from matplotlib import image
import seaborn as sns
import pandas as pd
```

```python
x_axis = [0,1,2]
fig,axes = plt.subplots(figsize = (10,8))
y_axis = [0.6668,0.3266,0.109]
plt.bar(x_axis,y_axis,facecolor = 'lightskyblue',width = 0.4)
plt.xlabel(r'arc_order (n)',fontsize = 14)
plt.ylabel('Accuracy Score (AVG)',fontsize = 14)
plt.title('Accuracy Score for MKM KPCA(arccosine) + LMNN+1 layer')
fig.savefig('Accuracy Score for MKM KPCA(arccosine) + LMNN+1 layer',bbox_inches = 'tight',dpi = 100)
```

```python
x_axis = range(6)
# fig,axes = plt.subplots(figsize = (10,8))
accu_score_0 = [0.827,0.828,0.835,0.845,0.827] #linear
accu_score_1 = [0.898,0.736,0.263,0.134,0.134] #poly
accu_score_2 = [0.097,0.106,0.102,0.091,0.099] #rbf
```

```python
# plot the accuracy score for arccosine kernel: 3 line 1 picture
x_axis = range(5)
fig,axes = plt.subplots(figsize = (10,8))
sns.lineplot(x = x_axis, y = accu_score_0,color = 'pink',label = 'linear')
sns.lineplot(x = x_axis, y = accu_score_1,color = 'hotpink',label = 'polynomial')
sns.lineplot(x = x_axis, y = accu_score_2,color = 'red',label = 'rbf')

plt.plot(x_axis,accu_score_0, 'yo')
plt.plot(x_axis,accu_score_1, 'yo')
plt.plot(x_axis,accu_score_2, 'yo')

plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.xlabel(r'Number of Layers(l)',fontsize = 14)
plt.ylabel('Accuracy Score (AVG)',fontsize = 14)
plt.legend(loc = 'upper right',prop={'size': 12})
plt.title('Accuracy Score for MKM KPCA + LMNN')
fig.savefig('Accuracy Score for MKM KPCA + LMNN',bbox_inches = 'tight',dpi = 100)
```
